Dealing with Missing Values

```python
#create a new DataFrame
sparse=pd.DataFrame(
    {
        'a':[1,2,3],
        'b':[np.nan,4,np.nan],
        'c':[5,np.nan,np.nan]
    }
)
```

```python
sparse
```

```python
#to check the null values,it returns the boolean values
sparse.isna()
```

```python
#we can fill the missing values with 0 as
sparse.fillna(0)
```

```python
sparse
```

```python
sparse.dropna(axis=1)
```
